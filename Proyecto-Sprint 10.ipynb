## Parte 2. An√°lisis del test A/B

````python
dates_groups = orders_filtered[['date', 'group']].drop_duplicates()
orders_daily = orders_filtered.groupby(['date', 'group']).agg({
    'transaction_id': pd.Series.nunique,
    'visitor_id': pd.Series.nunique,
    'revenue': 'sum'
}).reset_index().rename(columns={
    'transaction_id': 'orders',
    'visitor_id': 'visitors'
})
visits_daily = visitas.groupby(['date', 'group']).agg({'visits': 'sum'}).reset_index()
cumulative_data = pd.merge(orders_daily, visits_daily, on=['date', 'group'])
cumulative_data['revenue_cumulative'] = cumulative_data.groupby(['group'])['revenue'].cumsum()
cumulative_data['orders_cumulative'] = cumulative_data.groupby(['group'])['orders'].cumsum()
cumulative_data['visits_cumulative'] = cumulative_data.groupby(['group'])['visits'].cumsum()
````

````python
fig, ax = plt.subplots(figsize=(10, 6))
cumulative_data_a = cumulative_data[cumulative_data['group'] == 'A']
cumulative_data_b = cumulative_data[cumulative_data['group'] == 'B']
ax.plot(cumulative_data_a['date'], cumulative_data_a['revenue_cumulative'], label='Grupo A')
ax.plot(cumulative_data_b['date'], cumulative_data_b['revenue_cumulative'], label='Grupo B')
ax.set_title('Ingresos Acumulados por Grupo')
ax.set_xlabel('Fecha')
ax.set_ylabel('Ingresos Acumulados')
ax.legend()
plt.grid(True)
plt.show()
````

### El grupo B es el que m√°s ingresos ha acumulado subiendo de forma impactante alrededor del dia 17 y 21 de Agosto, mientras que el grupo A esta estable subiendo consecutiva.

````python
fig, ax = plt.subplots(figsize=(10, 6))
cumulative_data['avg_order_size_cumulative'] = cumulative_data['revenue_cumulative'] / cumulative_data['orders_cumulative']
ax.plot(cumulative_data[cumulative_data['group'] == 'A']['date'], cumulative_data[cumulative_data['group'] == 'A']['avg_order_size_cumulative'], label='Grupo A')
ax.plot(cumulative_data[cumulative_data['group'] == 'B']['date'], cumulative_data[cumulative_data['group'] == 'B']['avg_order_size_cumulative'], label='Grupo B')
ax.set_title('Tama√±o de Pedido Promedio Acumulado')
ax.set_xlabel('Fecha')
ax.set_ylabel('Tama√±o de Pedido Promedio')
ax.legend()
plt.grid(True)
plt.show()
````

### El gruoi A inicio arriba con pedidos acumulados y decayo drasticamente pero se recuper√≥, mientras que el grupo B tuvo una cantidad de pedidos acumulados un poco mayor a los 160 pero decayo levemente; pero esto nos da a conocer que hubo pedidos grandes.

````python
merged_cumulative_data = cumulative_data[cumulative_data['group'] == 'A'].merge(
    cumulative_data[cumulative_data['group'] == 'B'],
    left_on='date',
    right_on='date',
    how='left',
    suffixes=['_a', '_b']
)
merged_cumulative_data['relative_diff_size'] = (
    merged_cumulative_data['avg_order_size_cumulative_b'] / merged_cumulative_data['avg_order_size_cumulative_a'] - 1
)
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(merged_cumulative_data['date'], merged_cumulative_data['relative_diff_size'])
ax.axhline(y=0, color='black', linestyle='--')
ax.set_title('Diferencia Relativa en el Tama√±o de Pedido Promedio Acumulado (B vs A)')
ax.set_xlabel('Fecha')
ax.set_ylabel('Diferencia Relativa')
plt.grid(True)
plt.show()
````

### La gr√°fica se muestra volatil y existe una diferencia relativa alta ya que como se puede ver las estadisticas elevadas son la diferencia que hizo el grupo B.

````python
conversion_data = pd.merge(orders_daily, visits_daily, on=['date', 'group'])
conversion_data['conversion_rate'] = conversion_data['orders'] / conversion_data['visits']
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(conversion_data[conversion_data['group'] == 'A']['date'], conversion_data[conversion_data['group'] == 'A']['conversion_rate'], label='Grupo A')
ax.plot(conversion_data[conversion_data['group'] == 'B']['date'], conversion_data[conversion_data['group'] == 'B']['conversion_rate'], label='Grupo B')
ax.set_title('Tasa de Conversi√≥n Diaria')
ax.set_xlabel('Fecha')
ax.set_ylabel('Tasa de Conversi√≥n')
ax.legend()
plt.grid(True)
plt.show()
````

### La mayor parte de los precios estan por debajo de los 500, aquellas minorias que esten esten por encima de 830 son considerados anomalias.

````python
orders_by_users_a = pedidos[pedidos['group'] == 'A'].groupby('visitor_id')['transaction_id'].nunique().reset_index()
orders_by_users_b = pedidos[pedidos['group'] == 'B'].groupby('visitor_id')['transaction_id'].nunique().reset_index()
sample_a_orders = pd.concat([orders_by_users_a['transaction_id'], pd.Series(0, index=np.arange(visitas[visitas['group'] == 'A']['visits'].sum() - len(orders_by_users_a['transaction_id'])), name='transaction_id')], axis=0)
sample_b_orders = pd.concat([orders_by_users_b['transaction_id'], pd.Series(0, index=np.arange(visitas[visitas['group'] == 'B']['visits'].sum() - len(orders_by_users_b['transaction_id'])), name='transactionI_i')], axis=0)
p_value_conversion_raw = st.mannwhitneyu(sample_a_orders, sample_b_orders)[1]
alpha = 0.05
print(f"\n--- Significancia Estad√≠stica (Datos Brutos) ---")
print(f"P-value para la conversi√≥n: {p_value_conversion_raw:.3f}")
if p_value_conversion_raw < alpha:
    print("Conclusi√≥n: Existe una diferencia estad√≠sticamente significativa en la conversi√≥n.")
else:
    print("Conclusi√≥n: No hay una diferencia estad√≠sticamente significativa en la conversi√≥n.")
p_value_revenue_raw = st.mannwhitneyu(pedidos[pedidos['group'] == 'A']['revenue'], pedidos[pedidos['group'] == 'B']['revenue'])[1]
print(f"P-value para el tama√±o promedio de pedido: {p_value_revenue_raw:.3f}")
if p_value_revenue_raw < alpha:
    print("Conclusi√≥n: Existe una diferencia estad√≠sticamente significativa en el tama√±o promedio de pedido.")
else:
    print("Conclusi√≥n: No hay una diferencia estad√≠sticamente significativa en el tama√±o promedio de pedido.")
````

###  Tama√±o promedio de pedido: Esto es contradictorio con los gr√°ficos, lo que indica que los valores at√≠picos son tan extremos que sesgan el an√°lisis.
###  Conversi√≥n: Tuvo un impacto real en el n√∫mero de pedidos.

````python
orders_filtered_clean = orders_filtered[
    (orders_filtered['revenue'] <= 904) & (orders_filtered['visitor_id'].isin(orders_by_user[orders_by_user['orders'] <= 2]['visitor_id']))
]
orders_by_users_a_filtered = orders_filtered_clean[orders_filtered_clean['group'] == 'A'].groupby('visitor_id')['transaction_id'].nunique()
orders_by_users_b_filtered = orders_filtered_clean[orders_filtered_clean['group'] == 'B'].groupby('visitor_id')['transaction_id'].nunique()
sample_a_filtered = pd.concat([orders_by_users_a_filtered, pd.Series(0, index=np.arange(visitas[visitas['group'] == 'A']['visits'].sum() - len(orders_by_users_a_filtered)), name='orders')], axis=0)
sample_b_filtered = pd.concat([orders_by_users_b_filtered, pd.Series(0, index=np.arange(visitas[visitas['group'] == 'B']['visits'].sum() - len(orders_by_users_b_filtered)), name='orders')], axis=0)
p_value_conversion_filtered = st.mannwhitneyu(sample_a_filtered, sample_b_filtered)[1]
print(f"\n--- Significancia Estad√≠stica (Datos Filtrados) ---")
print(f"P-value para la conversi√≥n: {p_value_conversion_filtered:.3f}")
if p_value_conversion_filtered < alpha:
    print("Conclusi√≥n: Existe una diferencia estad√≠sticamente significativa en la conversi√≥n.")
else:
    print("Conclusi√≥n: No hay una diferencia estad√≠sticamente significativa en la conversi√≥n.")
p_value_revenue_filtered = st.mannwhitneyu(orders_filtered_clean[orders_filtered_clean['group'] == 'A']['revenue'], orders_filtered_clean[orders_filtered_clean['group'] == 'B']['revenue'])[1]
print(f"P-value para el tama√±o promedio de pedido: {p_value_revenue_filtered:.3f}")
if p_value_revenue_filtered < alpha:
    print("Conclusi√≥n: Existe una diferencia estad√≠sticamente significativa en el tama√±o promedio de pedido.")
else:
    print("Conclusi√≥n: No hay una diferencia estad√≠sticamente significativa en el tama√±o promedio de pedido.")
````
